```python
def non_norm(data):
    timeseries_list= list(set(data["timeseries"]))
    non_norm = []
    for timeseries in timeseries_list:
        if len(data[data['timeseries'] == timeseries]) != 36:
            non_norm.append(timeseries)
    return non_norm
```
